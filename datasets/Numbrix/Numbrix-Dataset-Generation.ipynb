```python
import numpy as np
import random
from pysat.solvers import Minisat22
import os
import pickle as pk
```

```python
def get_puzzle_data(data, n):
    solution = []
    puzz = []
    
    sol_idx = data.index("Solution") + 1
    for i in range(0, len(data[:sol_idx-2]), 4):
        line = data[i+2][1:-1].split('| ')
        row = [int(num.strip()) if num.strip().isdigit() else 0 for num in line]
        puzz.append(row)
        
    for line in data[sol_idx:]:
        row = [int(num) for num in line.split()]
        solution.append(row)
    
    return np.array(puzz), np.array(solution)
```

```python
path = "Numbrix6"
#path = "Numbrix5"

def read_puzzles(path, n=None):
    puzzles = []
    solutions = []
    dir_list = os.listdir(path)

    if n is None:
        num_puzz = len(dir_list)
    else:
        num_puzz = n
        
    for i in range(num_puzz):
        name = path + "\\" + dir_list[i]
        with open(name, 'r') as f:
            file_str = f.read()
        data = file_str.split('\n')
        data = list(filter(None, data))[5:]
        puzz, solution = get_puzzle_data(data, 5)
        puzzles.append(puzz), solutions.append(solution)
    return puzzles, solutions

puzzles, solutions = read_puzzles(path)
print(len(puzzles))
```

```python
def check_uniq(puzzles, solutions):
    puzz, sol = [], []
    uniq = {np.array2string(puzzles[i]): (puzzles[i], solutions[i]) for i in range(len(solutions))}
    for data in uniq.values():
        puzz.append(data[0])
        sol.append(data[1])
    return  puzz, sol # [array([1, 3, 2, 4]), array([1, 2, 3, 4])]
        
puzzles, solutions = check_uniq(puzzles, solutions)
print(len(puzzles))
```
